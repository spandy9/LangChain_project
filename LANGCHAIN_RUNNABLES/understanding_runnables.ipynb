```python
import random
```

```python
class NakliLLM:
    def __init__(self) -> None:
        print('NakliLLM initialized')

    def predict(self, prompt: str) -> dict[str, any]:
        response_list = [
            'NakliLLM response 1',
            'NakliLLM response 2',
            'NakliLLM response 3'
        ]
        return {
            'response': random.choice(response_list)
        }
```

```python
class NakliPromptTemplate:
    def __init__(self, template: str, input_variables: list[str]) -> None:
        self.template = template
        self.input_variables = input_variables

    def format(self, input_dict: dict[str, str]) -> str:
        return self.template.format(**input_dict)
```

### Runnable class build up using abstract class

```python
from abc import ABC, abstractmethod

class Runnable(ABC):
    @abstractmethod
    def invoke(input_data: any):
        pass
```

```python
class NakliLLM(Runnable):
    def __init__(self) -> None:
        print('NakliLLM initialized')

    def invoke(self, prompt: str) -> dict[str, any]:
        response_list = [
            'NakliLLM response 1',
            'NakliLLM response 2',
            'NakliLLM response 3'
        ]
        return {
            'response': random.choice(response_list)
        }
class NakliPromptTemplate(Runnable):
    def __init__(self, template: str, input_variables: list[str]) -> None:
        self.template = template
        self.input_variables = input_variables

    def invoke(self, input_dict: dict[str, str]) -> str:
        return self.template.format(**input_dict)

class NakliStrOutParser(Runnable):
    def __init__(self) -> None:
        pass
    def invoke(self, input_dict: dict[str: any]) -> str:
        return input_dict['response']
```

```python
class RunnableConnector(Runnable):
    def __init__(self, runnable_list: list[Runnable]) -> None:
        self.runnable_list = runnable_list
    def invoke(self, input_data: any) -> any:
        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)
        return input_data
```

```python
llm = NakliLLM()
prompt_template = NakliPromptTemplate(template='What is the capital of {country}?', input_variables=['country'])
parser = NakliStrOutParser()
chain = RunnableConnector([prompt_template, llm, parser])
response = chain.invoke({'country': 'India'})
print(response)
```
